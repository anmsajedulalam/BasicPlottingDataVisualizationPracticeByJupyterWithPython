```python
# JupyterPracticeBySajidWithPython
# Basic Plotting: 
# Importing pandas
import pandas as pd
# Importing pyplot for plotting from matplotlib library
from matplotlib import pyplot as plt
# Declaring three axes anb their values in three different arrays
x= [1, 2, 3]
y= [1, 4, 9]
z= [10,5,0]

# Firstly, plotting between x and y
plt.plot(x,y)
# Secondly, plotting between x and z
plt.plot(x,z)
# Giving a title to the plot
plt.title("test plot")
# Labelling the x axes
plt.xlabel("x")
# Labelling the y and z axes together
plt.ylabel("y and z")
# For showing the legend
plt.legend(["this is y", "this is z"])
# Finally, showing the plot
plt.show()
```
